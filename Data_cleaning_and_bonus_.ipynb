```python
NCVS=df_NCVS.copy()
NCVS=NCVS.drop(columns=['idper','yearq','race_ethnicity','newcrime','region','citizen','msa','race','hincome2','newoff','seriousviolent','notify',	'vicservices','marital','locality','weapcat','injury','serious','treatment','locality','locality','educatn2','veteran','locationr','offtracenew','newwgt'])
NCVS
```

```python
NCVS['ager'] = NCVS['ager'].replace([1, 2,3,4,5,6], ['12:17', '18:24', '25:34','35:49','50:64','>=65'])
NCVS['sex'] = NCVS['sex'].replace([1, 2], ['Male','Female'])
NCVS['hispanic'] = NCVS['hispanic'].replace([1, 2,88], ['Hispanic','Non-hispanic','residue'])
NCVS['hincome1'] = NCVS['hincome1'].replace([1, 2,3,4,5,6,7,88], ['Less than $7,500','$7,500 to $14,999','$15,000 to $24,999','$25,000 to $34,999','$35,000 to $49,999','$50,000 to $74,999','$75,000 or more','Unknown'])
NCVS['popsize'] = NCVS['popsize'].replace([-1,0,1,2,3,4,5], ['Invalid','Not a place','<100,000','100,000-249,999','250,000-499,999','500,000-999,999','>=1 million'])
NCVS['direl'] = NCVS['direl'].replace([1, 2,3,4,5,6],['Intimates','Other relatives','Well known/casual acquaintance','Strangers','Do not know relationship', 'Do not know number of offenders'])
NCVS['educatn1']=NCVS['educatn1'].replace([1,2,3,4,5,88],['No schooling','Grade school','Middle school','High school','College','Residue'])
NCVS['weapon']=NCVS['weapon'].replace([1,2,3],['Yes','No','Do not know'])
NCVS['offenderage']=NCVS['offenderage'].replace([1,2,3,4,5,88],['=<11','12:17','18-29','>=30','various ages offenders','Residue'])
NCVS['offendersex']=NCVS['offendersex'].replace([1,2,3,4,88],['Male','Female','Both','Unknown','Residue'])
NCVS['series']=NCVS['series'].replace([1,2],['Not a series crime','Series crime'])
NCVS
```

```python
NCVS.columns = NCVS.columns.str.replace('educatn1', 'Education level')
NCVS.columns = NCVS.columns.str.replace('direl', 'Victim-offender relationship')
NCVS.columns = NCVS.columns.str.replace('hincome1', 'Annual household income')
NCVS.columns = NCVS.columns.str.replace('wgtviccy', 'Victimization weight')
```

```python
#Train a machine/deep learning classifier to predict the likelihood of recidivism within 3 years of release based on the state of Georgia recidivism records.
from sklearn.ensemble import RandomForestClassifier
```

```python
Recidivism=df_Recidivism.copy()
```

```python
Recidivism.dropna(inplace=True)
```

```python

# import modules
import matplotlib.pyplot as mp
import seaborn as sb
  
  
# plotting correlation heatmap
dataplot=sb.heatmap(Recidivism.corr())
  
# displaying heatmap
mp.show()
```

```python
X=Recidivism.drop(columns=['Unnamed: 0'	,'id','recidivism_within_3years','recidivism_arrest_year1', 'recidivism_arrest_year2','recidivism_arrest_year3'])
X
```

```python
X=X.loc[:,(X.dtypes=='float').values | (X.dtypes=='bool').values]
X
```

```python
y=Recidivism.loc[:,['recidivism_within_3years']]
y
```

```python
# Splitting arrays or matrices into random train and test subsets
from sklearn.model_selection import train_test_split
# i.e. 70 % training dataset and 30 % test datasets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.30)
```

```python

# creating a RF classifier
clf = RandomForestClassifier(n_estimators = 1000)  

clf.fit(X_train, y_train)
  
# performing predictions on the test dataset
y_pred = clf.predict(X_test)
  
# metrics are used to find accuracy or error
from sklearn import metrics  
  
# using metrics module for accuracy calculation
print("ACCURACY OF THE MODEL: ", metrics.accuracy_score(y_test, y_pred))
```
